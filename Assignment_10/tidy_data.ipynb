```python
weather=pd.read_csv('https://raw.githubusercontent.com/ajitjadhav10/pydatadc_2018-tidy/master/data/weather.csv')
```

```python
weather
```

```python
# can use pd.melt or weather.melt

weather_melt=pd.melt(
    weather,
    id_vars=['id','year','month','element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)
```

```python
weather_tidy
```
